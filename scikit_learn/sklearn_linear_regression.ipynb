```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X_train = np.array([
    [10, 8, 2],
    [8, 6, 5],
    [12, 7, 1],
    [6, 6, 8],
    [15, 8, 0],
    [7, 5, 6],
    [9, 7, 3],
    [14, 9, 0],
    [5, 4, 10],
    [11, 8, 2]
])
y_train = np.array([85, 70, 90, 60, 95, 65, 80, 93, 50, 88])
```

```python
sgdr = LinearRegression()
sgdr.fit(X_train, y_train)
```

```python
w = sgdr.coef_
b = sgdr.intercept_

print("w, b using scikit    :", w, b)
print("w, b without scikit  : [ 0.46946335  1.12149526 -3.58354842] 78.67576652071745")
```

```python
y_hat = sgdr.predict(X_train)
y_hat_wo_scikit = np.array([85.18, 71.24, 88.58, 59.55, 94.69, 66.07, 80.00, 95.34, 49.67, 85.64])

for i in range(y_hat.shape[0]):
    print(f"prediction (w/ scikit): {y_hat[i]:0.2f}, prediction(w/o scikit): {y_hat_wo_scikit[i]}, target value: {y_train[i]}")
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(range(len(y_train)), y_train, color='blue', label='Target Value')
plt.scatter(range(len(y_hat)), y_hat, color='red', marker='x', label='Prediction')

for i in range(len(y_train)):
    plt.plot([i, i], [y_train[i], y_hat[i]], color='gray', linestyle='dashed')

plt.legend()
plt.show()
```

```python
X_test = np.array([[9, 4 ,5]])
y_pred = sgdr.predict(X_test)

print(f"Siswa dengan : \nwaktu belajar selama {X_test[0, 0]} jam per minggu, \nrata-rata waktu tidur selama {X_test[0, 1]} jam per hari, dan \ntidak hadir selama {X_test[0, 2]} hari dalam bulan, \ndiprediksi akan memperoleh nilai ujian sebesar {y_pred[0]:0.2f}")
```
